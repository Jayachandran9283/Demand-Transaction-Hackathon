```python
def KNN_predict(train_x,test_x,train_y,test_y,scaler,k_value,metric='manhattan',weights='uniform'):
    scaled_train=scaler.fit_transform(train_x,train_y)
    scaled_test=scaler.transform(test_x)
    KNN= KNeighborsRegressor(n_neighbors=k_value,weights=weights,metric=metric)
    KNN.fit(scaled_train,train_y)
    pred_train=KNN.predict(scaled_train)
    pred_test=KNN.predict(scaled_test)
    print(f'Rsquare of Train {round(KNN.score(scaled_train, train_y) , 2)}')
    print(f'Rsquare of Test {round(KNN.score(scaled_test, test_y) , 2)}')
    #print(f'MAE {metrics.mean_absolute_error(train_y, pred_train)}')
    return KNN_predict
```

```python
KNN_predict(train_x, test_x, train_y,test_y,StandardScaler(),1)
```

```python
for k in range(1, 11):
    print('R square on KNN using n_neighbors = {0}:'.format(k), end = ' ')
    KNN_predict(train_x, test_x, train_y,test_y,StandardScaler(), k)
```

```python
k=4
for metric in ['euclidean','minkowski','manhattan','chebyshev']:
    print('R square on {0} metrics \n'.format(metric),end=' ')
    KNN_predict(train_x, test_x, train_y,test_y, StandardScaler(), k,metric,weights='uniform')
```

```python
for weight in ['uniform','distance']:
    print('R square on {0} weights \n'.format(weight),end=' ')
    KNN_predict(train_x, test_x, train_y,test_y, StandardScaler(), k_value=4,metric='manhattan',weights=weight)
```
